```python
import pandas as pd

# Creating DataFrames
meal_info = pd.DataFrame({
    "Meal ID": [101, 102, 103, 104, 105],
    "Meal Type": ["Breakfast", "Lunch/Dinner", "Lunch/Dinner", "Snack", "Dessert"],
    "Recipe Name": ["Idli", "Hyderabadi Biryani", "Rajma Chawal", "Dhokla", "Gulab Jamun"],
    "Cuisine": ["South Indian", "Mughlai", "North Indian", "Gujarati", "North Indian"],
    "Time to Prepare (min)": [20, 60, 30, 20, 40],
    "Vegetarian": [True, False, True, True, True]
})

ingredients_info = pd.DataFrame({
    "Meal ID": [101, 101, 102, 102, 103, 104,106],
    "Ingredient ID": [201, 202, 203, 204, 205, 206,207],
    "Ingredient Name": ["Rice Batter", "Coconut Chutney", "Basmati Rice", "Chicken", "Kidney Beans (Rajma)", "Basmati Rice","Kela"],
    "Quantity (g/ml)": [500, 100, 750, 1000, 300, 500,12]
})

```

```python
meal_info
```

```python
ingredients_info
```

```python
#pd.merge(df1,df2,common="column name", hpw)
pd.merge(meal_info, ingredients_info, on="Meal ID", how="inner")
```

```python
pd.merge(meal_info, ingredients_info, on="Meal ID", how="left")
```

```python
pd.merge(meal_info, ingredients_info, on="Meal ID", how="right")
```

```python

# Merging DataFrames
pd.merge(meal_info, ingredients_info, on="Meal ID", how="outer")
```

When to Use:

When you need to combine datasets with shared keys (like Meal ID).
Supports different merge types: inner, left, right, and outer.

When to Use:

When you want to stack DataFrames (vertically by default).
Use axis=1 for horizontal concatenation.

When to Use:

When both DataFrames have meaningful indices.
Use join() when working with indices for alignment.

```python
import pandas as pd

# Orders DataFrame
orders = pd.DataFrame({
    "Order ID": [1001, 1002, 1003, 1004],
    "Product ID": ["P001", "P002", "P003", "P001"],
    "Customer Name": ["John Doe", "Jane Smith", "Alice Brown", "Bob White"],
    "Quantity Ordered": [2, 5, 1, 3],
    "Region": ["North America", "Europe", "Asia", "South America"]
})

# Products DataFrame
products = pd.DataFrame({
    "Product Code": ["P001", "P002", "P003", "P004"],
    "Product Name": ["Smartphone", "Laptop", "Headphones", "Tablet"],
    "Price (USD)": [500, 1000, 150, 300],
    "Region": ["North America", "Europe", "Asia", "South America"]
})
```

```python
orders
```

```python
products
```

```python

# Merging based on multiple keys
merged_data = pd.merge(
    orders,
    products,
    left_on=["Product ID", "Region"],
    right_on=["Product Code", "Region"],
    how="inner"
)

merged_data
```
