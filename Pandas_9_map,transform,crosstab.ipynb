1. Mapping: map()
Definition:
map() is used to apply a function or mapping to each element of a Series. It works on a single column or row of the dataset.

Parameters:

Function: A function, lambda expression, or callable to apply to each element.
Dictionary: A mapping dictionary to replace values in the Series.
Series: Another Series for element-wise mapping.

```python
regional_mapping = {"Andhra Pradesh": "South Indian"}

df["Regional Classification"] = df["Cuisine"].map(regional_mapping)

df.head()
```

2. Mapping on DataFrame: applymap()
Definition:
applymap() is used to apply a function element-wise across an entire DataFrame. It works on each individual element.

Parameters:

A callable function to apply to each element of the DataFrame.
When to Use:
When you need to transform or clean all elements of a DataFrame, such as formatting or applying a calculation to every cell.

Example Use Case:
Suppose the dataset has string values, and you want to convert all elements to lowercase for uniformity.

```python
# Converting all string elements to lowercase
df_lower = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df_lower.head()
```

```python
import pandas as pd

# Example Dataset
data = {"Category": ["A", "A", "B", "B", "C", "C"], "Value": [10, 20, 30, 40, 50, 60]}
df2 = pd.DataFrame(data)

df2
```

```python
df2.groupby("Category")["Value"].mean()
```

```python
# Applying transform()
df2.groupby("Category")["Value"].transform("mean")

```

```python
#transform() is designed to broadcast aggregated results to each row of the group.
#This is particularly useful in scenarios where group-level statistics (e.g., mean, sum, rank)
#are needed in their original context.

```

```python
df2.groupby("Category")["Value"].groups
```

```python
df2.groupby("Category")["Value"].get_group("A")
```

```python
df2.groupby("Category")["Value"].get_group("A").std()
```
