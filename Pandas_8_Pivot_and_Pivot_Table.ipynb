```python
import pandas as pd

# Use the raw GitHub URL
#url = "https://raw.githubusercontent.com/AkashRajpuria/Pandas-Session/main/Datasets_for_all_the_Sessions/Indori.xlsx"
url = "https://raw.githubusercontent.com/AkashRajpuria/Pandas-Session/main/Datasets_for_all_the_Sessions/Modified_data.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

 Pivot
The pivot function reshapes data to produce a new DataFrame. It requires the dataset to have unique index/column combinations (i.e., no duplicates).

```python
#DataFrame.pivot(index=None, columns=None, values=None)
df.pivot(index='Cuisine', columns='Type', values='avg_cooking_time_in_minutes')
```

```python
grouped = df.groupby(['Cuisine', 'Type'])['avg_cooking_time_in_minutes'].mean().reset_index()
grouped
```

```python
grouped.pivot(index='Cuisine', columns='Type', values='avg_cooking_time_in_minutes')
```

# Pivot Table

```python
#DataFrame.pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None)
```

```python
df.pivot_table(index='Cuisine', columns='Type', values='avg_cooking_time_in_minutes', aggfunc='mean')
```

```python
df.head(2)
```

```python
df.pivot_table(index='Cuisine', columns='Type', values='Ingredient Count', aggfunc='mean')
```

```python
# When to Use Which?
# Use pivot if your dataset has unique combinations of index and columns and you do not need to aggregate.
# Use pivot_table if your dataset has duplicates or you need aggregation.
```

Find average cooking time for each cuisine and type



```python
df.pivot_table(index='Cuisine', columns='Type',values='avg_cooking_time_in_minutes',  aggfunc='mean')

```

```python
#Count the number of recipes for each type (Vegetarian/Non-Vegetarian) by cuisine:

```

```python
df.pivot_table(values='Recipe Name', index='Cuisine', columns='Type', aggfunc='count', fill_value=0)
```

```python
df.pivot_table(values='Recipe Name', index='Type', columns='Cuisine', aggfunc='count', fill_value=0)
```

Identify the maximum ingredient count for each cuisine and type:


```python
df.pivot_table(values='Ingredient Count', index='Cuisine', columns='Type', aggfunc='max')
```

Combine multiple aggregation functions (mean and sum) for cooking time by cuisine:


```python
df.pivot_table(values='avg_cooking_time_in_minutes', index='Cuisine', columns='Type', aggfunc=['mean', 'sum'])
```

Add custom index levels by combining multiple columns:



```python
df.pivot_table(values='avg_cooking_time_in_minutes', index=['Cuisine', 'Type'], columns='Meal', aggfunc='mean',fill_value="Not supplied")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AkashRajpuria/Pandas-Session/refs/heads/main/Datasets_for_all_the_Sessions/Retail.csv")
df.head()
```

```python
#Analyze Total Sales by Region Using pivot_table, we can calculate the total sales for each region:
pivot_sales = df.pivot_table(values='Sales', index='Region', aggfunc='sum')
pivot_sales
```

```python
#Profitability Analysis by Category and Region Pivot data to see average profits per category across regions:
pivot_profit = df.pivot_table(values='Profit', index='Category', columns='Region', aggfunc='mean', fill_value=0)
pivot_profit
```

```python
#Business Insight: Electronics is the most profitable in the North and East regions, while Furniture is exclusive to the West.
```

```python
#Track Quantity Sold by Sub-Category and Region Pivot the dataset to see quantities sold for each sub-category in each region:

pivot_quantity = df.pivot_table(values='Quantity', index='Sub-Category', columns='Region', aggfunc='sum', fill_value=0)
pivot_quantity
```

```python
#Performance Over Time: Sales by Date and Region Pivot data to view daily sales across regions:

pivot_time = df.pivot_table(values='Sales', index='Date', columns='Region', aggfunc='sum', fill_value=0)
pivot_time
```

```python
#Combine Multiple Aggregations Use multiple aggregation functions (sum and mean) to analyze sales and profits

pivot_agg = df.pivot_table(values=['Sales', 'Profit'], index='Region', aggfunc=['sum', 'mean'])
pivot_agg
```
