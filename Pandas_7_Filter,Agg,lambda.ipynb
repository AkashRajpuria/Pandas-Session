Aggregating Data: aggregate()
The aggregate() function is used to apply one or more aggregation operations on columns of a DataFrame. Aggregation combines data into a summary form, such as calculating the sum, mean, count, or any custom aggregation for each group or entire DataFrame.

```python
aggregated_data = df.groupby('Type').aggregate({
    'avg_cooking_time_in_minutes': 'mean',
    'Ingredient Count': 'sum'
})
```

```python
aggregated_data
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Cuisine': ['Andhra Pradesh', 'Andhra Pradesh', 'Andhra Pradesh', 'Andhra Pradesh', 'Andhra Pradesh'],
    'Type': ['Vegetarian', 'Non-Vegetarian', 'Non-Vegetarian', 'Vegetarian', 'Vegetarian'],
    'Time to Prepare (min)': [23, 35.5, 23, 18, 2160],
    'Ingredient Count': [13, 20, 24, 12, 8]
}
df2 = pd.DataFrame(data)

df2
```

```python
# Aggregating the average preparation time and total ingredient count by type
aggregated_data = df2.groupby('Type').aggregate({
    'Time to Prepare (min)': 'mean',
    'Ingredient Count': 'sum'
})
print(aggregated_data)
```

 Filtering Data: filter()
The filter() function is used to subset rows or groups of a DataFrame based on a condition.

Key Points
Can be applied to rows or groups.
Useful for keeping only specific rows that meet a condition.

```python
df['avg_cooking_time_in_minutes'] < 30
```

```python
filtered_data = df[df['avg_cooking_time_in_minutes'] < 30]
filtered_data
```

```python
# Grouping by Cuisine and filtering groups with total Ingredient Count > 20
filtered_groups = df.groupby('Cuisine').filter(lambda x: x['Ingredient Count'].sum() > 5000)
filtered_groups
```

```python
filtered_groups['Cuisine'].value_counts()
```

```python
filtered_groups.groupby("Cuisine").ngroups
```

```python
filtered_groups.groupby("Cuisine").groups
```

```python
# Filtering groups with average preparation time < 100
filtered_avg_time = df.groupby('Cuisine').filter(lambda x: x['avg_cooking_time_in_minutes'].mean() < 100)
filtered_avg_time
```

```python
# Adding a column for word count in Cuisine using lambda
df['Cuisine Word Count'] = df['Cuisine'].apply(lambda x: len(x.split()))
df[['Cuisine', 'Cuisine Word Count']]
```

Challenging Questions
1. Find the total ingredient count for each cuisine type and filter those with more than 30 ingredients.

```python
# Grouping by Cuisine and summing Ingredient Count
ingredient_sum = df.groupby('Cuisine').aggregate({'Ingredient Count': 'sum'})

# Filtering cuisines with total ingredient count > 30
filtered_cuisines = ingredient_sum[ingredient_sum['Ingredient Count'] > 30]
filtered_cuisines
```

Which type of recipe (Vegetarian/Non-Vegetarian) has the longest average preparation time?

```python
# Calculating average preparation time for each type
average_time = df.groupby('Type').aggregate({'avg_cooking_time_in_minutes': 'mean'})

average_time
```

```python
average_time['avg_cooking_time_in_minutes']
```

```python

# Getting the type with the longest average preparation time
longest_avg_type = average_time.loc[average_time['avg_cooking_time_in_minutes'] == average_time['avg_cooking_time_in_minutes'].max()]
longest_avg_type
```

. Create a new column indicating if the preparation time is above or below average for the entire dataset.

```python
overall_avg_time = df['avg_cooking_time_in_minutes'].mean()
overall_avg_time
```

```python
# Calculate the overall average preparation time

# Add a new column using lambda
df['Above Average Prep Time'] = df['avg_cooking_time_in_minutes'].apply(lambda x: 'Above Average' if x > overall_avg_time else 'Below Average')
df[["Recipe Name",'avg_cooking_time_in_minutes', 'Above Average Prep Time']]
```
