```python
def avg_time(duration):
  #print(duration)
  list_duration  = duration.split("-")
  if len(list_duration)>1:

    if "hour" in list_duration[1]:

      end = list_duration[1].replace(" hours","").replace(" hour","")
      start = float(list_duration[0])
      start = float(start)*60
      end = float(end)*60

      return (start+end)/2

    if "minute" in list_duration[1]:
      end = list_duration[1].replace(" minutes\n","").replace(" minutes","").replace(" minute","")
      start = float(list_duration[0])
      end = float(end)
      start = float(start)
      return (start+end)/2


    if "day" in list_duration[1]:

      end = list_duration[1].replace(" days","").replace("days","").replace(" day","").replace("day","")
      start = float(list_duration[0])
      end = float(end)*1440
      start = float(start)* 1440
      return (start+end)/2


  else:
    return 1440 *3
```

The apply() function is used to apply a custom or built-in function to a column or row. For example:

We want to calculate the word count in the "Recipe Name" column.

Using groupby()
The groupby() function is used to group data based on one or more columns. For instance:

We group recipes by "Type" and calculate the average preparation time.

```python
grouped = recipe_data.groupby('Type')['avg_cooking_time_in_minutes'].mean()
```

```python
# Step 1: Define a function to count the number of ingredients
def count_ingredients(ingredients):
    return len(ingredients.split(',')) if pd.notna(ingredients) else 0

# Step 2: Apply the function to count ingredients for each recipe
recipe_data['Ingredient Count'] = recipe_data['Ingredients'].apply(count_ingredients)
```

```python
recipe_data.head()
```

```python

# Step 3: Group by 'Cuisine' and 'Type' to calculate the average ingredient count
grouped_ingredients = recipe_data.groupby(['Cuisine', 'Type'])['Ingredient Count'].mean()

grouped_ingredients
```

```python
grouped_ingredients.reset_index()
```

```python
# Convert the result to a DataFrame for better visualization
ingredient_analysis = grouped_ingredients.reset_index().rename(columns={'Ingredient Count': 'Avg Ingredient Count'})

ingredient_analysis.sort_values(by='Avg Ingredient Count', ascending=False).head()
```
