```python
def avg_time(duration):
  #print(duration)
  list_duration  = duration.split("-")
  if len(list_duration)>1:

    if "hour" in list_duration[1]:

      end = list_duration[1].replace(" hours","").replace(" hour","")
      start = float(list_duration[0])
      start = float(start)*60
      end = float(end)*60

      return (start+end)/2

    if "minute" in list_duration[1]:
      end = list_duration[1].replace(" minutes\n","").replace(" minutes","").replace(" minute","")
      start = float(list_duration[0])
      end = float(end)
      start = float(start)
      return (start+end)/2


    if "day" in list_duration[1]:

      end = list_duration[1].replace(" days","").replace("days","").replace(" day","").replace("day","")
      start = float(list_duration[0])
      end = float(end)*1440
      start = float(start)* 1440
      return (start+end)/2


  else:
    return 1440 *3
```

```python
from collections import Counter

# Function to split and flatten the ingredients into individual items
def split_ingredients(ingredient_list):
    # Remove any brackets and split by commas
    return [ingredient.strip() for ingredient in re.sub(r'[\[\]]', '', ingredient_list).split(',')]

# Apply the function to the 'Ingredients' column to get a list of individual ingredients
recipe_data['Individual Ingredients'] = recipe_data['Ingredients'].apply(split_ingredients)

# Flatten the ingredients and count frequencies by Cuisine and Type
def most_common_ingredient(group):
    ingredients = sum(group, [])
    counter = Counter(ingredients)
    return counter.most_common(1)[0] if counter else (None, 0)

# Group by 'Cuisine' and 'Type', and calculate the most frequently occurring ingredient
most_common_ingredients = recipe_data.groupby(['Cuisine', 'Type'])['Individual Ingredients'].apply(most_common_ingredient).reset_index()
most_common_ingredients[['Most Common Ingredient', 'Count']] = pd.DataFrame(most_common_ingredients['Individual Ingredients'].tolist(), index=most_common_ingredients.index)
most_common_ingredients.drop('Individual Ingredients', axis=1, inplace=True)

most_common_ingredients
```

## Longest Recipe Names by Cuisine and Type


Question:
Which cuisines are known for long or descriptive recipe names, and how do vegetarian and non-vegetarian dishes compare in terms of name length?

Objective:

Use apply() to calculate the length of each recipe name (character or word count).
Group by Cuisine and Type to find the longest average name length.
Highlight cuisines with particularly descriptive or concise naming conventions

```python
# Function to calculate the length of each recipe name by character count
def recipe_name_length(name):
    return len(name)

# Alternatively, for word count
# def recipe_name_length(name):
#     return len(name.split())

# Apply the function to the 'Recipe Name' column to get the name length
recipe_data['Name Length'] = recipe_data['Recipe Name'].apply(recipe_name_length)

# Group by 'Cuisine' and 'Type', then calculate the average name length
avg_name_length = recipe_data.groupby(['Cuisine', 'Type'])['Name Length'].mean().reset_index()

# Identify cuisines with the longest and shortest average recipe name lengths
longest_names = avg_name_length.nlargest(1, 'Name Length')
shortest_names = avg_name_length.nsmallest(1, 'Name Length')

(avg_name_length, longest_names, shortest_names)
```
